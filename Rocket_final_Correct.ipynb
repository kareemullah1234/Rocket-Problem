```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline
```

```python
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = '16'
```

```python
g=9.81

rho = 1.091

a = math.pi * 0.5**2

CD =0.15

ve = 325
```

```python
T = 25

dt = 0.1

N = int(T/dt) + 1

t = numpy.linspace(0,T,num=N)

print('t=',t)

#print(t[50])
```

```python

mp = numpy.zeros(N)

mp[:50] =20

print('mp is \n',mp)


v0 = 0.0




u = numpy.array([0])

#u = numpy.empty(N)


print('u is\n',u)
```

```python
#for i,j in zip(t,mp):
u_values = [0]
for n in range(1, N):
        rhs = -g + (mp[n] * ve)/(150 - (mp[n] * t[n])) - ((rho * u* abs(u) *a *CD)/(2*(150 - mp[n] * t[n])))
        print('rhs=',rhs)
        
        print('mp[n] is =',mp[n])
        
        print('t[n] is=',t[n])
        
        print('n=',n)
        
        #print('rhs shape',rhs.shape) 
        u = u + (dt * rhs)
        u_values.append(u)
        print('u=',u)
        print('--------------------------------------------------------------------')
        #h = u[0]
#print('h is',h)

print(u_values)
```

```python
pyplot.plot(t,u_values)
```
